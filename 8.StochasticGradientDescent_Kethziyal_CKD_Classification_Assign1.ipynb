```python
import pandas as pd
import numpy as npy
import matplotlib as pyplt
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset['classification_yes'].value_counts()
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
independent
```

```python
dependent=dataset[['classification_yes']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import SGDClassifier
param_grid={'loss':['hinge','modified_huber', 'squared_hinge', 'perceptron', 
                    'squared_error', 'huber', 'epsilon_insensitive', 'squared_epsilon_insensitive'],
         'penalty':['l2', 'l1', 'elasticnet']} 
grid=GridSearchCV(SGDClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,Y_train)
```

```python
grid_predictions=grid.predict(X_test)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,grid_predictions)

from sklearn.metrics import classification_report
clf_report=classification_report(Y_test,grid_predictions)
```

```python
print(cm)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(Y_test,grid_predictions,average='weighted')
print("f1_macro values for the best parameter{}:".format(grid.best_params_),f1_macro)
```

```python
re=grid.cv_results_
print(re)
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
Y_prob = grid.predict(X_test)
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(Y_test, Y_prob)
roc_auc = auc(fpr, tpr)
```

```python
print("ROC_AUC for StochasticGD Model:",roc_auc)
```
